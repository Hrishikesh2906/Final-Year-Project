```python
# Function to preprocess text data
def preprocess_text(text):
    # Remove punctuations
    text = text.translate(str.maketrans('', '', string.punctuation))

    # Remove unwanted words (http, co, amp and \n)
    text = text.replace('http', '').replace('co', '').replace('amp', '').replace('\n', '')
    
    # Return processed text
    return text
```

```python
def preprocess_tweets(df):
    print('Preprocessing tweets DataFrame...')

    # Drop duplicates based on tweetId and reset index for DataFrame
    df.drop_duplicates(subset='tweetId', inplace=True)
    df.reset_index(drop=True, inplace=True)

    # Drop records with null value for "source" column
    df.dropna(subset=['source'], inplace=True)

    # Convert "date" column to datetime and keep only date
    df['date'] = pd.to_datetime(df['date'], format='%Y/%m/%d').dt.date

    # Preprocess text in "renderedContent" column
    print('\nPreprocessing text in "renderedContent" field...')
    df['textContent'] = df.progress_apply(lambda x: preprocess_text(x['renderedContent']), axis=1)

    # Extract device type from "source" column
    print('\nExtracting device type from "source" field...')
    df['deviceType'] = df.progress_apply(lambda x: x['source'].split()[1].split('/')[-1][0:-1], axis=1)

    print('\ntweets DataFrame transformed successfully!')
    print('  - Duplicate records dropped based on tweet ID')
    print('  - Records where source field contains null value dropped')
    print('  - Date column converted to datetime')
    print('  - Text content preprocessed to remove punctuations and unwanted words')
    print('  - Extracted device type from source column')
    print('\n\nNumber of rows after preprocessing: {}\nNumber of columns after preprocessing: {}\n'.format(newData.shape[0], newData.shape[1]))
    return
```

```python
# Function to preprocess users data
def preprocess_users(df):
    print('Preprocessing users DataFrame...')

    # Drop duplicate tweets based on tweetId and reset index for DataFrame
    users.drop_duplicates(subset='userId', inplace=True)
    users.reset_index(drop=True, inplace=True)

    # Convert "created" column to datetime and keep only date
    df['created'] = pd.to_datetime(df['created'], format='%Y/%m/%d').dt.date

    print('\nusers DataFrame transformed successfully!')
    print('  - Duplicate records dropped based on user ID')
    print('  - Date of profile creation converted to datetime')
    print('\n\nNumber of rows after preprocessing: {}\nNumber of columns after preprocessing: {}\n'.format(users.shape[0], users.shape[1]))
    return
```

```python
from wordcloud import WordCloud

# Text data from tweets DataFrame
tweet_text = ' '.join(text for text in newData.textContent.astype(str))

# Create wordcloud from tweet text
wc = WordCloud(background_color="white")
wordcloud = wc.generate(tweet_text)

# Display wordcloud
plt.axis('off')
plt.figure(figsize=(16, 10))
plt.imshow(wordcloud)
plt.show()
```
